To test regularized differentiation scheme of Rick Chartrand 
https://github.com/rickchartrand/regularized_differentiation

```python
import numpy as np
import matplotlib.pyplot as plt
import regularized_differentiation.regularized_gradient
import regularized_differentiation.differentiation
```

```python
x = np.linspace(0, 6*np.pi,10000)
y = np.sin(x)
mean_absolute_v = np.mean(np.abs(y))
y_noisy = y + np.random.normal(loc=0.0, scale=0.1*mean_absolute_v,size=len(x))
```

```python
plt.plot(x,y_noisy, label='noisy')
plt.plot(x,y, label='clean')
plt.legend()
```

```python
firstD_clean = np.gradient(y,x)
firstD_noisy = np.gradient(y_noisy, x)
secondD_clean = np.gradient(firstD_clean, x)
secondD_noisy = np.gradient(firstD_noisy, x)
```

```python
y_twod = np.expand_dims(y_noisy,axis=1)
```

```python
y_twod = np.repeat(y_twod,5,axis=1)
```

```python
firstD_tvr = regularized_differentiation.regularized_gradient.tv_regularized_gradient(
    y_twod,tuning_parameter=0.33, splitting_parameter=1.5e-3, iterations=3000, diagnostics=True)
firstD_tvr = firstD_tvr/(x[1]-x[0])

firstD_tvr = firstD_tvr[1,:,0]
tt = np.expand_dims(firstD_tvr, axis=1)
tt = np.repeat(tt,5,axis=1)

secondD_tvr = regularized_differentiation.regularized_gradient.tv_regularized_gradient(
    tt,tuning_parameter=0.33, splitting_parameter=1.5e-3, iterations=3000, diagnostics=True)
secondD_tvr = secondD_tvr/(x[1]-x[0])
secondD_tvr = secondD_tvr[1,:,0]
```

```python
plt.plot(x,firstD_tvr)
plt.xlim(np.min(x),np.max(x))
plt.ylim(-1,1)
```

```python
plt.plot(x, secondD_tvr)
plt.plot(x, secondD_clean)
plt.ylim(-1,1)
```
